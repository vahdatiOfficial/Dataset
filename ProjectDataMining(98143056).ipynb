```python
import os
import tarfile
from zipfile import ZipFile
import urllib
```

```python
DOWNLOAD_BASE = r'https://github.com/vahdatiOfficial/Dataset/archive/refs/heads/'
DOWNLOAD_URL = DOWNLOAD_BASE + 'main.zip'
DATASET_PATH = os.path.join(".","horse-colic")
print(DOWNLOAD_BASE)
print(DOWNLOAD_URL)
print(DATASET_PATH)

```

```python
def fetch_horse_colic_data(url=DOWNLOAD_URL,path=DATASET_PATH):
    #os.makedirs(path,exist_ok=True)
    zip_path = os.path.join('.','main.zip')
    urllib.request.urlretrieve(url,zip_path)
    horse_colic_zip = ZipFile(zip_path)
    horse_colic_zip.extractall(path=path)
    horse_colic_zip.close()
```

```python
def load_horse_colic_data(path=DATASET_PATH):
    path = path + '\Dataset-main'
    csv_path = os.path.join(path,'horse-colic.csv')
    return pd.read_csv(csv_path)
```

```python
from six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
```

```python
def treeImage(tree):
    dot_data = StringIO()
    export_graphviz(tree, out_file=dot_data, filled=True, rounded=True, special_characters=True)
    graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
    return Image(graph.create_png())
```

```python
treeImage(tree_gini)
```
